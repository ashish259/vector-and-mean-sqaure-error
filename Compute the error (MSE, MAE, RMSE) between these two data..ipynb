```python
#calculate mse, mae, rmse
#mse
from sklearn.metrics import mean_squared_error

#calculation for single column of  the dataframe,
x_true = d1['Avg_Pit_Depth_mm']  #true calue
x_pred = d2['Avg_Pit_Depth_mm']  #calculated value

# Calculation of Mean Squared Error (MSE) 
print('Mean Sq. Error(MSE): ',mean_squared_error(x_true,x_pred)) 

#calculating rmse(rmse)
print('Root mean Sq. Error(RMSE): ', (mean_squared_error(x_true,x_pred)) ** 0.5)
```
